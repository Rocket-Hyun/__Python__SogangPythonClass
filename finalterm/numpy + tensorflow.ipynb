```python
import numpy as np
import tensorflow as tf
tf.set_random_seed(777)  # for reproducibility
```

```python
commit_array = np.genfromtxt('./data/github_commits.csv', delimiter=',')[1:,1:]
```

```python
commit_array.tolist()
```

```python
commit_array[0]
```

```python

#평균 구하기
commit_mean = np.mean(commit_array, axis=1)
print("기존 차원: ", commit_mean.shape)

# array 차원 변경
commit_mean = np.reshape(commit_mean, (39,1))
print("변경된 차원: ",commit_mean.shape)
```

```python
commit_mean[:3]
```

```python
# Try to find value for W and b to compute y_data = x_data * W + b  
# We know that W should be 1 and b should be 0
# But let's TensorFlow figure it out 
W = tf.Variable(tf.random_normal([6,1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
#### Now we can use X and Y in place of x_data and y_data
#### placeholders for a tensor that will be always fed using feed_dict
#### See http://stackoverflow.com/questions/36693740/
X = tf.placeholder(tf.float32, shape=[None, 6])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
# Out hypothesis XW+b
#hypothesis = X * W + b
# 여러 차원 끼리 비교일 때
hypothesis = tf.matmul(X, W) + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001)
train = optimizer.minimize(cost)
```

```python
# Launch the graph in a session.
sess = tf.Session()
# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())
```

```python
# for step in range(10001):
#     cost_val, W_val, b_val, _ = \
#         sess.run([cost, W, b, train],
#                  feed_dict={X: commit_array, Y:commit_mean})
#     if step % 1000 == 0:
#         print(step, cost_val, W_val, b_val)
        

for step in range(10001):
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train], feed_dict={X: commit_array, Y: commit_mean})
    if step % 1000 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
```
