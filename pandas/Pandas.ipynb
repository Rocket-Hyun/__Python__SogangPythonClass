```python
areas = data.area1.unique()
mean_wages = []

for ar in areas:
    mean_wages.append(data.loc[data['area1'] == ar][['hourly_wage']].mean().hourly_wage)
    
df = pd.DataFrame(data={'평균 시급':mean_wages}, index=areas, columns=['평균 시급'])
df.sort_values(by='평균 시급', ascending=False)
```
